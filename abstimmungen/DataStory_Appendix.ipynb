## Landwirtschaft

Die Bev철lkerungsdichte pro km짼 und die Siedlungsf채che k철nnten beides Indikatoren f체r urbane oder l채ndliche Gemeinden sein. Um diesen Zusammenahng zu untersuchen, k철nnen wir die beiden Datens채tze zusammen anschauen und auch mit einem Scatter Plot visualisieren.


```python
desc = manager.joined_data.groupby("Cluster (k=9)")[["Bev철lkerungs-dichte pro km짼", "Siedlungsfl채che in %"]].median().sort_values("Siedlungsfl채che in %")
display(desc)

scatter = plt.scatter(
        manager.joined_data["Bev철lkerungs-dichte pro km짼"],
        manager.joined_data["Siedlungsfl채che in %"],
        c=manager.joined_data[f"Cluster (k={9})"],
        cmap="tab20",
    )
plt.grid()
plt.title("Gemeinden Stadt-Land")
plt.xlabel("Bev철lkerungs-dichte pro km짼", wrap=True)
plt.ylabel("Siedlungsfl채che in %", wrap=True)
handles, labels = scatter.legend_elements(prop="colors")
plt.legend(handles, labels, title="Cluster")
plt.show()

scatter = plt.scatter(
        desc["Bev철lkerungs-dichte pro km짼"],
        desc["Siedlungsfl채che in %"],
        c=desc.reset_index()[f"Cluster (k={9})"],
        cmap="tab20",
    )

plt.grid()
plt.title("Stadt-Land Gruppen")
plt.xlabel("Bev철lkerungs-dichte pro km짼", wrap=True)
plt.ylabel("Siedlungsfl채che in %", wrap=True)
handles, labels = scatter.legend_elements(prop="colors")
plt.legend(handles, labels, title="Cluster")
plt.show()
```

Der Zusammenhang zwischen der Bev철lkerungsdichte und der Siedlungsfl채che ist in den Clustern sehr deutlich ersichtlich.
