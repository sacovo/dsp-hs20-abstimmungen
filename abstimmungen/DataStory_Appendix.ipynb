## Landwirtschaft

Die Bevölkerungsdichte pro km² und die Siedlungsfäche könnten beides Indikatoren für urbane oder ländliche Gemeinden sein. Um diesen Zusammenahng zu untersuchen, können wir die beiden Datensätze zusammen anschauen und auch mit einem Scatter Plot visualisieren.


```python
desc = manager.joined_data.groupby("Cluster (k=9)")[["Bevölkerungs-dichte pro km²", "Siedlungsfläche in %"]].median().sort_values("Siedlungsfläche in %")
display(desc)

scatter = plt.scatter(
        manager.joined_data["Bevölkerungs-dichte pro km²"],
        manager.joined_data["Siedlungsfläche in %"],
        c=manager.joined_data[f"Cluster (k={9})"],
        cmap="tab20",
    )
plt.grid()
plt.title("Gemeinden Stadt-Land")
plt.xlabel("Bevölkerungs-dichte pro km²", wrap=True)
plt.ylabel("Siedlungsfläche in %", wrap=True)
handles, labels = scatter.legend_elements(prop="colors")
plt.legend(handles, labels, title="Cluster")
plt.show()

scatter = plt.scatter(
        desc["Bevölkerungs-dichte pro km²"],
        desc["Siedlungsfläche in %"],
        c=desc.reset_index()[f"Cluster (k={9})"],
        cmap="tab20",
    )

plt.grid()
plt.title("Stadt-Land Gruppen")
plt.xlabel("Bevölkerungs-dichte pro km²", wrap=True)
plt.ylabel("Siedlungsfläche in %", wrap=True)
handles, labels = scatter.legend_elements(prop="colors")
plt.legend(handles, labels, title="Cluster")
plt.show()
```

Der Zusammenhang zwischen der Bevölkerungsdichte und der Siedlungsfläche ist in den Clustern sehr deutlich ersichtlich.
